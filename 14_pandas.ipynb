## Merge data

```python
import pandas as pd
data1 = {
    "name": ["sally","marry","John"],
    "age": [77,40,30]
}

data2 = {
    "name": ["sally","Peter","michy"],
    "gender":['F','M','M']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

newdf = pd.merge(df1,df2, how='outer')
print(newdf)
newdf = pd.merge(df1,df2, how='inner')
print(newdf)
newdf = pd.merge(df1,df2, how='left')
print(newdf)
newdf = pd.merge(df1,df2, how='right')
print(newdf)
```

```python
df1 = pd.DataFrame({'name': ['digonto','prottoy','ruhul'],
                   'dept': ['management','physics','agriculture'],
                   'material status': ['Unmarried','unmarried','unmarried']})

df2 = pd.DataFrame({'name': ['digonto','prottoy2','ruhul2'],
                   'age': ['24','25','26'],
                   'cgpa': ['3.00','3.20','3.60']})

print(df1)
print(df2)
newdf2 = pd.merge(df1,df2,on="name")
print(newdf2)
```

```python
newdf3 = pd.merge(df1,df2,how="outer")
print(newdf3)
```

## Fill null value

```python
newdf3['cgpa']=newdf3['cgpa'].fillna(2.75)
print(newdf3['cgpa'])
```

```python
newdf3.age = newdf3.age.fillna(30)
print(newdf3.age)
```

## Convert object type into int64

```python
newdf3['cgpa']=newdf3['cgpa'].astype('str').astype('float')
print(newdf3['cgpa'])
```

```python
newdf3['age']=newdf3['age'].astype('str').astype('int')
newdf3.age
```

## Double the value

```python
f = lambda x: x*2
newdf3['age'] = newdf3['age'].apply(f)
print(newdf3.age)
```

## Sorting

```python
newdf3.sort_values(by='age',ascending = False)
```

## Filtering

```python
#Filter records with age more than 25
filter1 = newdf3['age'] > 25
#Applyfilter to dataframe
filtered_new = newdf3[filter1]
#View filtered dataframe
print(filtered_new)
```
