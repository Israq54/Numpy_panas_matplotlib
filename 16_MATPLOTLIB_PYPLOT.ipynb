```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
#changing the figure
fig = plt.figure(figsize=(10,5))

#changing the datapoints
plt.plot(x,y,linewidth=.50,linestyle=':',color='blue',marker='o')
plt.title("Line plot demo")
plt.xlabel("x-axis")
plt.ylabel("y_axis")
plt.legend(['line1'],loc='best')
#to build grid
plt.grid(True)
plt.show()
```

## plot multiple plot

```python
 import numpy as np
 import matplotlib.pyplot as plt

 x = np.arange(0,10,0.1)
 y1 = 2*x+5
 y2 = x*x

 #create two or more plot in  fig
 plt.subplot(1,3,1)
 #plotting the points
 plt.plot(x,y1)
 plt.title('Graph1')

 plt.subplot(1,3,2)
 plt.plot(x,y2)
 plt.title('Graph2')

 plt.subplot(1,3,3)
 plt.plot(x,y2)
 plt.title('Graph3')

 plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10,0.1)
y1 = x*x*x
y2 = np.sqrt(x)

plt.subplot(1,3,1)
plt.plot(x,y1)
plt.title('graph1')
plt.grid(True)

plt.subplot(1,3,2)
plt.plot(x,y1)
plt.title('Graph2')
plt.grid(True)

plt.subplot(1,3,3)
plt.plot(x,y2)
plt.title('Graph3')
plt.grid(True)

plt.show()
```

## Create a bar plot [it has catagorical data]

```python
#creating the dataset
data = {'apple': 50,
        'lemon': 20,
        'banana': 1}

names = list(data.keys())
value = list(data.values())

fig = plt.figure(figsize = (10,5))
#creating the bar plot
plt.bar(names,value,color = 'r')
plt.show()
```

## Customizing the scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#creating the dataset
a = [10,20,30,40,50,60,70]
b = [1,2,3,4,0,8,6]
x = [4,5,6,1,2,3,8]

fig = plt.figure(figsize = (10,10))

plt.scatter(a,b, color = 'green', s = 200, edgecolors = 'red', marker = 'o', alpha = 0.9)
plt.scatter(a,x, color = 'red', s = 200, edgecolor = 'blue', marker = '4', alpha = 1)
plt.scatter(b,x,color = 'blue', s =200)

plt.legend(['b','x','a'],loc='upper left')

plt.title("Scatter plot")
plt.xlabel("Xaxis")
plt.ylabel("y axis")
plt.grid(True)
plt.savefig('scater plot.png')
plt.show()
```

## How to create a box plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

#data preparation
total = [20,4,1,30,20,40,30]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,5,10,20,50,60,20,45]

data = list([total, order, discount])
#ploting the data
plt.boxplot(data,showmeans = True)
plt.title("Box plot")
plt.grid(True)
plt.show()
```

##area plot

```python
#creating the dataset
data = {'apple': 50,
        'lemon': 20,
        'banana': 1}

names = list(data.keys())
value = list(data.values())

fig = plt.figure(figsize = (10,5))
#creating the bar plot
plt.stackplot(names,value,color = 'green',alpha=0.1)
plt.plot(names,value,color='black')
plt.legend(['names','values'],loc='best')
plt.show()
```
